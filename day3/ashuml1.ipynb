```python
import pandas as pd 
import matplotlib.pyplot as plt 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
# loading data using pandas df 
df = pd.read_csv("https://raw.githubusercontent.com/redashu/Datasets/refs/heads/master/advertising.csv")
```

```python
# checking metda
df.info()
```

```python
# top 5 rows 
df.head()
```

```python
# choosing features and labels -- input and output 
labels = df['Sales']
features = df[['TV','Radio','Newspaper']]
#print(labels)
#print(features)

# Spliting data into Training and testing 
training_features,test_features,training_labels,test_labels=train_test_split(features,labels,test_size=0.1)
```

```python
# lets calling Linear Regression Model 
model = LinearRegression()
# lets give Training data -- 90% features and labels
trained_model=model.fit(training_features,training_labels)
```

```python
# pass model to testing features see the outcome 
predicted_output=trained_model.predict(test_features)
print(predicted_output)
```

```python
# checking actual output 
print(test_labels)
```

```python
# ploting scatter of actual vs predicted output 
plt.xlabel("actual sales")
plt.ylabel("predicted sales")
plt.title("Advertising sales company : Actual vs predicted")
plt.scatter(test_labels,predicted_output)
plt.plot([min(test_labels),max(test_labels)],[min(predicted_output),max(predicted_output)],color='red',linestyle='--')
plt.show()
```

```python
# Analyze the impact of each features on sales by using coef values
coef_values = trained_model.coef_
feature_name = features.columns
print(feature_name)
```

```python
# print coeff 
print("Impact of adv options on sales")
for feature,coef in zip(feature_name,coef_values):
    print(f"{feature}: {coef:.4f}")
```

```python
# ploting this 
plt.bar(feature_name,coef_values,color=['blue','green','red'])
plt.xlabel('adv options')
plt.ylabel('ceoff values')
plt.title("Impact of adv option on sales")
plt.show()
```
